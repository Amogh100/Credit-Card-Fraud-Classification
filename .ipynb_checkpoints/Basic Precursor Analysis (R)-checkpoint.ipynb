```r
creditCardData <- read.csv('creditcard.csv')
```

```r
length(which(creditCardData$Class == 0))/nrow(creditCardData)
```

```r
#Test the normality assumption for the columns
```

```r
library("dplyr")
```

```r
relevantData <- creditCardData %>% select(-one_of("Class"))
```

```r
sampleRelevantData <- sample_n(relevantData, 5000)
```

```r
lapply(sampleRelevantData, shapiro.test)
#For all columns the null hypothesis that the column data 
#is distributed normally is rejected.
#It'll be interesting to see how the Gaussian Naive Bayes Classifier
#works
```
